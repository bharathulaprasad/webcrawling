## Web data mining using Pandas


```python
import pandas as pd
import requests
import io
from urllib.request import urlopen,Request
url='https://www.worldometers.info/coronavirus/'
content=requests.get(url).content
c=pd.read_html(io.StringIO(content.decode('utf-8')),displayed_only=False,header = 0)
#print(c)
raw=c[0]
raw
raw.shape
```
